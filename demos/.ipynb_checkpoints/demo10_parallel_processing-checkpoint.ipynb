```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from multiprocessing import Pool
import time

from orpl.synthetic import gen_synthetic_spectrum
```

# Parallel spectrum processing

The notebook shows a short demo on how to use the builtin *multiprocessing* library to do parallel processing of Raman spectra on all your cpu cores. Lets start by creating a pandas dataframe that contains a bunch of synthetic spectrum.

```python
# Creation of dataframe
df = pd.DataFrame(columns=['PID', 'Raw', 'r_target', 'b_target', 'noise'])


nspectrum = 1000 # lets generate 1000 spectrum

for i in range(nspectrum):
    s, r, b, n = gen_synthetic_spectrum(preset='tylenol',
                                    rb_ratio=np.random.rand(),
                                    noise_level=np.random.rand()/50)
    df.at[i, 'PID'] = 'Random patient # {}'.format(i)
    df.at[i, 'Raw'] = s
    df.at[i, 'r_target'] = r
    df.at[i, 'b_target'] = b
    df.at[i, 'noise'] = n
    
    
# I add 2 new columns to the dataframe now to avoid confusion later
# these lines will hold the processed raman (r_) and baseline (b_)
# spectrum
df['r_'] = object
df['b_'] = object

df.head()
```

### Processing function
 
I define a processing function to work on a line of the dataframe. Doing it this way allows us to use all columns for processing. Maybe fields other than *Raw* are relevant? (it's not the case here, but could be). The function returns a processed raman and baseline spectrum in a tuple (r_, b_) instead of filling the dataframe cells right away. (this is necessary for the multiprocessing pool).

```python
from orpl.baseline_removal import bubblefill

def process_spectrum(dfline):
    '''
        Process spectrum s in a dfline
        dfline = df.iloc[i]
    '''
    s = dfline.Raw # the Raw spectrum from a df line
    r_, b_ = bubblefill(s) # baseline removal
    return r_, b_
```

### Singlethreaded processing

We can test our processing function in a single threaded loop.

```python
start = time.time()

results = [] # we first add the results to a list and add them to the df after.

# processing
for i in range(len(df)):
    r_, b_ = process_spectrum(df.iloc[i])
    results.append((r_, b_))
    
    
# adding results to df
for i in range(len(df)):
    df.at[i, 'r_'] = results[i][0]
    df.at[i, 'b_'] = results[i][1]
    
stop = time.time()
print('Singlethreaded duration = {:.3} s'.format(stop-start))
```

### Multithreaded processing

Now let's run the multithreaded version

#### Works and Tested on LINUX | NOT on windows | mac ???

```python
# Let's reset the df columns for r_ and b_ (flushes the results from the single threaded loop)
df['r_'] = object
df['b_'] = object

start = time.time()
ncore = 8 # the number of cores we want to use for // computing

# Multi threaded execution
p = Pool(ncore) # this creates a pool of cpu worker

# creates all workers to use the process_spectrum with a list of all df lines
results = p.map(process_spectrum, [df.iloc[i] for i in range(len(df))])

p.close() # this closes the worker pool so python waits until it is done

# adding results to df
for i in range(len(df)):
    df.at[i, 'r_'] = results[i][0]
    df.at[i, 'b_'] = results[i][1]

stop = time.time()
print('Multithreaded duration = {:.3} s'.format(stop-start))
```

### Multithreaded processing

Now let's run the multithreaded version

#### Works and Tested on LINUX | windows ??? | mac ???

```python
# Let's reset the df columns for r_ and b_ (flushes the results from the single threaded loop)
df['r_'] = object
df['b_'] = object

start = time.time()
ncore = 8 # the number of cores we want to use for // computing

# Multi threaded execution
p = Pool(ncore) # this creates a pool of cpu worker

# creates all workers to use the process_spectrum with a list of all df lines
results = p.map_async(process_spectrum, [df.iloc[i] for i in range(len(df))])
results = results.get()
p.close() # this closes the worker pool so python waits until it is done

# adding results to df
for i in range(len(df)):
    df.at[i, 'r_'] = results[i][0]
    df.at[i, 'b_'] = results[i][1]

stop = time.time()
print('Multithreaded duration = {:.3} s'.format(stop-start))
```

### Validation the processing worked

Running this cell plots a random spectrum from the dataframe and the baseline and raman computed by the multithreaded processing

```python
i = np.random.randint(0, len(df))

plt.plot(df.Raw[i], label = 'Raw spectrum')
plt.plot(df.r_[i], label = 'Processed Raman')
plt.plot(df.b_[i], label = 'Processed Baseline')
plt.legend()
```
