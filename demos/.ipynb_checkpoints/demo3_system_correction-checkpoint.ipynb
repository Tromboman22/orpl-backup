```python
import json

import numpy as np
import matplotlib.pyplot as plt
```

synthetic# System response correction

The system correction curve can be computed by the `getCorrectionCurve()` function using a NIST standard spectrum.

This notebook shows the procedure to apply the system response correction on the first acquisition of the `bacon.json` experiment.

### Loading the data

```python
# Loading NIST data
with open('data/samples/bacon/standard.json', encoding='utf8') as file:
    nist=json.load(file)
nist = nist[10] # keeping only the 10th acquisition
plt.plot(nist['RawSpectra']);
```

```python
# Loading the bacon data
with open('data/samples/bacon/bacon.json', encoding='utf8') as file:
    bacon=json.load(file)
bacon = bacon[1]
plt.plot(bacon['RawSpectra']);
```

### Cropping the spectra

```python
crop_pos = 50 # where to crop the spectra

# Cropping nist
nist['cropped'] = np.array(nist['RawSpectra'])
nist['cropped'] = nist['cropped'][crop_pos:, :]

# Cropping bacon
bacon['cropped'] = np.array(bacon['RawSpectra'])
bacon['cropped'] = bacon['cropped'][crop_pos:, :]

# Plotting results
plt.figure(figsize=[10, 7])
plt.subplot(2,1,1)
plt.title('cropped NIST')
plt.plot(nist['cropped']);
plt.subplot(2,1,2)
plt.title('cropped bacon')
plt.plot(bacon['cropped']);
```

### Computing system response

```python
from orpl.calibration import compute_irf

# computing system response from nist mean spectrum
sys_response = compute_irf(nist['cropped'].mean(axis=1))

# plotting system response
plt.plot(sys_response)
plt.ylim([0,1.1])
```

### Applying system response on bacon

```python
# applying system correction
bacon['corrected'] = bacon['cropped'].transpose()/sys_response
bacon['corrected'] = bacon['corrected'].transpose()

# plotting
plt.plot(bacon['corrected']);
plt.title('System corrected Bacon spectra')
```
