# Synthetic spectra demo 1 : Presets

This notebook shows hows to generate synthetic spectra using presets from the *synthetic.py* library.
A comparaison is also made between synthetic spectra and the experimental version that they are attempting to mimic.

```python
import numpy as np
import matplotlib.pyplot as plt

from synthetic import gen_synthetic_spectrum
```

```python
# Define plot function (so I dont put this everywhere)
def plot_things(exp, syn):
    plt.figure(figsize=[20,4])
    plt.subplot(1, 3, 1)
    plt.title('Fluorescence')
    plt.plot(exp['f']/exp['f'].max(), label='experimental')
    plt.plot(syn['f']/syn['f'].max(), label='synthetic', linewidth=0.8)
    plt.grid()
    plt.legend()
    plt.subplot(1, 3, 2)
    plt.title('Raman')
    plt.plot(exp['r']/exp['r'].max(), label='experimental')
    plt.plot(syn['r']/syn['r'].max(), label='synthetic', linewidth=0.8)
    plt.grid()
    plt.legend()
    plt.subplot(1, 3, 3)
    plt.title('Total')
    plt.plot(exp['s']/exp['s'].max(), label='experimental')
    plt.plot(syn['s'], label='synthetic', linewidth=0.8)
    plt.grid()
    plt.legend()
    
```

## Nylon preset

Comparaison between the nylon preset and an experimental spectrum measured on pure nylon.

```python
# Loading experimental (*REAL*) nylon spectrum
data = np.genfromtxt(open('samples/nylon.csv'), delimiter=',')
exp = dict()
exp['f'] = data[1:, 1] # Fluorescence
exp['r'] = data[1:, 2] # Raman
exp['s'] = exp['f'] + exp['r'] # Total (fluo+raman+noise)
```

```python
# Generate synthetic spectrum from nylon preset
syn = dict()
syn['s'], syn['r'], syn['f'], syn['n'] = gen_synthetic_spectrum(preset='nylon', 
                                                                raman_flu_ratio=0.45, 
                                                                noise_level=0)
```

```python
# Plotting experimental vs synthetic
plot_things(exp, syn)
plt.savefig('nylon.png', dpi=320)
```
